```python
crime_data = pd.read_csv("MergedData.csv")
crime_data.head()
```

```python
crime_data["Crm Cd Desc"] = crime_data["Crm Cd Desc"].map(d1)
```

```python
crime_data["Crm Cd Desc"].isnull().sum()
```

```python
areaCrimeCounts = crime_data.groupby("AREA NAME").size().reset_index(name="Crime Count")
print(areaCrimeCounts)
```

```python
rowswithNA = crime_data[crime_data.isna().any(axis=1)]
```

```python
area_crime_counts = areaCrimeCounts.sort_values(by="Crime Count", ascending = False)
```

```python
area_crime_counts["Crime Count"]
print(area_crime_counts)
```

```python
bins = [0, 1000, 3294, 4766, 6241]
labels = ["Low Danger", "Medium Danger", "High Danger", "Very High Danger"]
```

```python
# Define bins and corresponding danger levels
bins = [0, 1000, 3000, 5000, float("inf")]
labels = ["Low Danger", "Medium Danger", "High Danger", "Very High Danger"]

# Assign Danger Levels based on Crime Count
area_crime_counts["Danger Level"] = pd.cut(area_crime_counts["Crime Count"], bins=bins, labels=labels, right=False)

# Print the updated DataFrame
print(area_crime_counts)
```

```python
# Map Danger Levels from area_crime_counts to crime_data
danger_level_mapping = area_crime_counts.set_index("AREA NAME")["Danger Level"].to_dict()

# Add the Danger Level column to crime_data based on AREA NAME
crime_data["Danger Level"] = crime_data["AREA NAME"].map(danger_level_mapping)

# Display the updated crime_data
print(crime_data.head())
```
